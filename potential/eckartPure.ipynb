```python
import sympy
height,width,energy,mass,hbar=sympy.symbols('H,W,E_\\text{eigen},m,hbar',positive=True)
variable=sympy.symbols('ell',real=True)
position=sympy.symbols('x',real=True)
waveFunction=sympy.symbols('Psi',cls=sympy.Function)
potential=height*sympy.sech(width*position)**2
schrodinger=-hbar**2/2/mass*waveFunction(position).diff(position,2)+(potential-energy)*waveFunction(position)
sympy.Eq(schrodinger,0)
```

```python
schrodinger=schrodinger.rewrite(sympy.exp).simplify()
sympy.Eq(schrodinger,0)
```

```python
x=next(iter(next(filter(lambda node:node!=sympy.S.Reals,sympy.solveset(sympy.exp(2*width*position)+variable,position,domain=sympy.S.Reals).args))))
x
```

```python
derivativePosition2=sympy.diff(waveFunction(variable).diff(variable)/x.diff(variable),variable)/x.diff(variable)
derivativePosition2
```

```python
schrodinger=sympy.expand_mul(schrodinger.subs(waveFunction(position).diff(position,2),derivativePosition2).subs(waveFunction(position),waveFunction(variable)).subs(position,x))
schrodinger=sympy.use(schrodinger,lambda node:node/schrodinger.coeff(waveFunction(variable).diff(variable,2)),level=1).collect(waveFunction(variable))
sympy.Eq(schrodinger,0)
```

```python
schrodinger.subs(schrodinger.coeff(waveFunction(variable))*waveFunction(variable),schrodinger.coeff(waveFunction(variable)).apart(variable)*waveFunction(variable))
```

```python
schrodinger=schrodinger.subs(schrodinger.coeff(waveFunction(variable))*waveFunction(variable),
                             sympy.apart(schrodinger.coeff(waveFunction(variable))*variable*(variable-1),variable)*waveFunction(variable)/variable/(variable-1))
sympy.Eq(schrodinger,0)
```

```python
characteristicExponent=sympy.symbols('characteristicExponent')
characteristicExponent=list(
    map(lambda denominator:sympy.solve((characteristicExponent+schrodinger.coeff(waveFunction(variable).diff(variable)).coeff(1/denominator[1])-1)*characteristicExponent
                                       +(-1)**(denominator[0]+1)*schrodinger.coeff(waveFunction(variable)/variable/(variable-1)).coeff(1/denominator[1]),characteristicExponent),
        enumerate([variable,variable-1]))
)+[sympy.solve((characteristicExponent+1-schrodinger.coeff(waveFunction(variable).diff(variable)).subs(variable-1,1).subs(variable,1))*characteristicExponent
               +next(filter(lambda node:not node.has(variable),schrodinger.coeff(waveFunction(variable)/variable/(variable-1)).args)),characteristicExponent)]
characteristicExponent
```

```python
solution=variable**characteristicExponent[0][0]*(variable-1)**characteristicExponent[1][0]*sympy.hyper(
    [sum(map(lambda characteristicExponent:characteristicExponent[0],characteristicExponent)),
     sum(map(lambda characteristicExponent:characteristicExponent[0],characteristicExponent[:-1]))+characteristicExponent[-1][-1]],
    [1+characteristicExponent[0][0]-characteristicExponent[0][-1]],variable)
solution
```

```python
hypergeometric=sympy.symbols('Phi',cls=sympy.Function)(variable)
hypergeometricEquation=sympy.expand(schrodinger.subs(waveFunction(variable),hypergeometric*variable**characteristicExponent[0][0]*(variable-1)**characteristicExponent[1][0])
                                    .doit()/variable**characteristicExponent[0][0]/(variable-1)**characteristicExponent[1][0]).collect(hypergeometric)
hypergeometricEquation=hypergeometricEquation.subs(hypergeometricEquation.coeff(hypergeometric),hypergeometricEquation.coeff(hypergeometric).cancel().factor()).subs(
    hypergeometricEquation.coeff(hypergeometric.diff(variable)),hypergeometricEquation.coeff(hypergeometric.diff(variable)).apart(variable,full=True))
sympy.Eq(hypergeometricEquation,0)
```

```python
hyper=next(filter(lambda node:isinstance(node,sympy.hyper),solution.args))
assert(not (1-hyper.bq[0]+sum(hyper.ap))-hypergeometricEquation.coeff(hypergeometric.diff(variable)).coeff(1/(variable-1)))
assert(not sympy.simplify(hyper.bq[0]-hypergeometricEquation.coeff(hypergeometric.diff(variable)).coeff(1/variable)))
assert(not sympy.simplify(sympy.prod(hyper.ap)-hypergeometricEquation.coeff(hypergeometric)*variable*(variable-1)))
solution=[solution,solution.subs(hyper,variable**(1-hyper.bq[0])*sympy.hyper([hyper.ap[0]-hyper.bq[0]+1,hyper.ap[1]-hyper.bq[0]+1],[2-hyper.bq[0]],variable)),
          solution.subs(hyper,(-variable)**-hyper.ap[0]*sympy.hyper([hyper.ap[0],hyper.ap[0]-hyper.bq[0]+1],[hyper.ap[0]-hyper.ap[1]+1],1/variable)).powsimp(),
          solution.subs(hyper,(-variable)**-hyper.ap[1]*sympy.hyper([hyper.ap[1],hyper.ap[1]-hyper.bq[0]+1],[hyper.ap[1]-hyper.ap[0]+1],1/variable)).powsimp()]
solution
```

```python
solution=list(map(lambda node:node.subs(variable,-sympy.exp(2*width*position)),solution))
solution
```

```python
solution=solution[:2]+list(map(lambda solution:sympy.use(sympy.use(solution,
                                                                   lambda node:sympy.expand(node.base/sympy.exp(2*width*position))**node.exp*sympy.exp(2*width*position*node.exp) 
                                                                   if node.has(-sympy.exp(2*width*position)-1) else node,level=1).powsimp(),
                                                         lambda node:sympy.expand_mul(node) if isinstance(node,sympy.exp) else node,level=1),solution[-2:]))
solution
```

```python
solution=list(map(lambda solution:sympy.use(solution,lambda node:(-node.base)**node.exp*(-1)**node.exp if isinstance(node,sympy.Pow) else node,level=1).powsimp(),solution))
solution
```

```python
coefficient=[sympy.gamma(1-hyper.bq[0])*sympy.gamma(hyper.ap[0]-hyper.ap[1]+1)/sympy.gamma(hyper.ap[0]-hyper.bq[0]+1)/sympy.gamma(1-hyper.ap[1]),
             sympy.gamma(hyper.bq[0]-1)*sympy.gamma(hyper.ap[0]-hyper.ap[1]+1)/sympy.gamma(hyper.ap[0])/sympy.gamma(hyper.bq[0]-hyper.ap[1])*sympy.exp((hyper.bq[0]-1)*sympy.pi*sympy.I)]
import numpy
reactant=numpy.dot(coefficient,solution[:2])
reactant
```

```python
reactant=sympy.use(reactant,lambda node:(-1)**(node.args[0]/sympy.pi/sympy.I) if isinstance(node,sympy.exp) and not node.has(position) else node,level=2).powsimp()
magnitude=next(filter(lambda node:isinstance(node,sympy.Pow) and node.has(-1),solution[2].args))
reactant=sympy.use(reactant,lambda node:sympy.powsimp(node/magnitude),level=1)
reactant
```

```python
product=sympy.powsimp(solution[2]/magnitude)
product
```

```python
reflection=1/sympy.prod(filter(lambda node:node.has(sympy.gamma),next(filter(lambda node:not node.has(-sympy.I*position),reactant.args)).args))
reactant=sympy.use(reactant,lambda node:node*reflection,level=1)
reactant
```
